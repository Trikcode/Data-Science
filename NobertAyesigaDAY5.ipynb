```python
import requests
import pandas as pd

url = 'https://www.xeno-canto.org/api/2/recordings?query=cnt:uganda'
headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
response = requests.request("GET", url, headers=headers, data={})
jsondata = response.json()

recordings = jsondata['recordings']
df = pd.DataFrame(recordings)
df.to_csv('uganda.csv', index=False)


```

```python
#webscrapp  html page
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://keithgalli.github.io/web-scraping/webpage.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
# print(soup.prettify())
#load only web page not html
# url3 = 'https://jeffgeoff.com';
# page3 = requests.get(url3)
# soup3 = BeautifulSoup(page3.content, 'lxml')
# print(soup3.prettify())
```

```python

table2 = soup.select('table.hockey-stats')[0]
columns = table2.select('thead th')
column_names = [c.string for c in columns]
print(column_names)


```

```python
table_rows = table2.select('tbody tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    # row = [tr.text for tr in td]
    row2 = [str(tr.get_text()).strip() for tr in td]
    l.append(row2)
df = pd.DataFrame(l, columns=column_names)
df.head()
# print(df)
```
